```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import urllib
import requests
```

```python
path = r"...\texas.csv"
texas = pd.read_csv(path, encoding="latin1")
```

```python
stopwords = set(STOPWORDS)
```

```python
deathtext = " ".join(texas.loc[texas["LastStatement"] != "None", "LastStatement"])
```

```python
path2 = r"...\skull.jpg"

mask = np.array(Image.open(path2))
```

```python
image_colors = ImageColorGenerator(mask)
```

```python
wordcloud = WordCloud(stopwords=stopwords, max_words = 10000, mask = mask, width=2000, height=2000, color_func=image_colors,
                     background_color = "white")
wordcloud.generate(deathtext)

plt.figure(figsize=(20,10))
plt.imshow(wordcloud, interpolation='bilinear')

plt.axis("off")
plt.show()
```
